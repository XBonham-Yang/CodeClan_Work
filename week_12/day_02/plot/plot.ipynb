# Data Viz 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd 
demo_data = pd.read_csv("plot/data/life_expectancy_and_income.csv")
```

```python
demo_data_2019 = (
    demo_data
    .loc[demo_data.year == 2019]
    .copy()
)
```

```python
plt.scatter(x='income_per_person',
            y = 'life_expectancy',
            data=demo_data_2019)
plt.xlabel('Income')
plt.ylabel('Life exp')
plt.title('life vs income')
# set limits 
plt.xlim(0,80000)
plt.ylim(0,100)
# controling tick marks 
plt.yticks(list(range(0,110,10)))
#just shows the plot not rendering code 
plt.show()
```

```python
#histograms 
#plt.hist()
```

```python
plt.hist(x='income_per_person', data = demo_data_2019)
plt.xlabel('Income$')
plt.ylabel('count')
plt.title('dist income')
plt.show()
```

Plot a boxplot showing the distribution of income per person for 2019

```python
plt.boxplot(x='income_per_person',vert = False, data = demo_data_2019)
plt.xlabel('Income$')
plt.ylabel('count')
plt.title('dist income')
plt.show()
```
